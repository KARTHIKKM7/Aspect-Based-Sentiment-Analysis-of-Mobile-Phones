```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

# Sample reviews
reviews = reviews = [
    "very good product at this price",
    "bad performance",
"within 2 months entire display is gone filled with line phone is very slow",
"worth of money keep buying",
"don't buy this product",
"in moto g 62 phone design, display, battery  all things is good but main problem  in my phone is network,  any sim i try jio and airtel but network problem , sometime  no network is come at the same time in orher phone full network is come. if you have any solutions  pl. provide  me. ..",
"the phone back camera is weste and mobile is all right",
"i'm giving this review after a month of use and i think at this price you may get good performance in realme phones or poco phone.",
"worst",
"slower than a turtle.....it doesn't work properly....don't waste your money on this guys.",
"a very bad product",
"apart from camera it's power pack.",
"worst mobile",
"bad product if you want slow than go for it",
"very bad mobile camera quality very bad",
"low display quality",
"display problem ahara",
"good one",
"lag",
"original quality.",
"bad performance and very slow phone",
"touch screen not working",
"okok",
"waste products",
"not fulfilling expectation as per budget.",
"good",
"classic in this segment",
"not really good at this price go for other",
"very bad product  can't expect from motorola always hanging problem.touch response veery slow touch now it's working 2 -4 days later.",
"motorola g22 mast beautiful look",
"problems right after 1 year of purchase",
"the processor is too slow, camera quality is average.i will not advise anyone to buy this phone as it is a painful experience to wait for the basic function to start as the phone is too slow",
"not as expected ",
"very bad phone",
"software is very slow in this mobile",
"horrible product",
"worst phone i have ever seen very poor performance one app takes 5-10 sec to open very lag luster phone",
"worst product. don't buy. lags too much.",
"good mobile in affordable price",
"awesome product...",
"worst product ever .i suggest everyone not to purchase phones of motorola very bad performance",
"don't like",
"don't buy this smartphone its very poor performance.",
"hang prblm",
"superb",
"ok very nice",
"no comments",
"very nice looking so very good battery",
"slow processor",
"poor",
"i am very disappointed with this phone , this is very slow , when i want to click a photo in a car  camra opened after 15 to 20 sec , so i want to sale this...please suggest i am right and wrong",
"phone good but hanging problem",
"very worst product totally wast for money",
"mobile waste of money not working properly phone hanging",
"display not tach",
"the slowest phone i have ever used, don't even think for this phone",
"do not.buy the phone.i had taken it got damaged 2 times within 1 year now it is not working.",
"product is good to use. i am using this phone for more than a year.",
"best phone in low range",
"third class",
"very bad ,charging time heated  and  camera  is  no  50  mp  ...... lagging issue  .. dont purchase",
"okk product",
"not bad",
"lacking problem",
"nice phone",
"not best",
"worst experience ever. dont buy any moto phone.",
"very low quality",
"the phone hangs a lot.even a day to day task on this phone becomes very tedious",
"please don't buy this phone.it is the slowest phone ever",
"you should regret after buying it.",
"to good all",
"after a month this hang too much. don't buy it, total waste of money.",
"performance not good of processer",
"not at all satisfied with this product the phone is too much hang from the day i have purchased",
"super",
"not happy",
"product is not good",
"ok",
"performance not good all time hang problem and i am  not satisfy moto phones very poor quality",
"not bad product",
"mobile hangs a lot fast doesn't work",
"super",
"very bad products",
"sometimes it get hang",
"writing my review after using this mobile for 4 months it is getting restarted automatically 4-5 times a day for the last few days.sometimes it goes to flight mode when i receive a call.sometimes the phone rings but there is no option to receive the call.will write more soon",
"very very good mobile",
"ok with budget.",
"superb",
"overall phone processor and performance very good problem only with late delevery with flipkart everytime",
"it's good n design is amazing ",
"new phone hang problem",
"very nice my pasand",
"a very good camera phone in budget range with excellent battery and  plastic case is within the box company shoud mention it",
"nice",
"worst phone ever on performance",
"camera quality is not good",
"camera quality is too low",
"nan",
"phone hang",
"on this price segment this phone is best....",
"this is very bad product, don't waste your money bought this but it have hanging problem and no return policy supported under 10 days",
"good looking phone, reported brand, pretty much like a stock android however there is some visible lag without any add on app and basic usage.",
"good ha",
"this is bad device",
"motorola mobile is bad no battery backup only hang hang hang",
"value for money battery performance are superb",
"good",
"good nice",
"very nice, i love it",
"awesome mobile thank you flipkart",
"looking",
"nice performance and happiness to get these mobile"
]

# Initialize the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Aspects to analyze
aspects_to_analyze = ["camera", "price", "battery", "performance","display","moto"]

# Initialize a dictionary to store sentiment scores for each aspect
aspect_sentiments = {aspect: [] for aspect in aspects_to_analyze}

# Perform aspect-based sentiment analysis
for review in reviews:
    for aspect in aspects_to_analyze:
        if aspect in review.lower():
            sentiment_score = sia.polarity_scores(review)
            aspect_sentiments[aspect].append(sentiment_score['compound'])

# Display the sentiment scores for each aspect
for aspect, scores in aspect_sentiments.items():
    print(f"Aspect: {aspect}")
    for i, score in enumerate(scores, 1):
        print(f"Review {i}: Sentiment Score = {score:.2f}")
    print()
```

```python
average_aspect_sentiments = {}
for aspect, sentiment_scores in aspect_sentiments.items():
    if sentiment_scores:
        average_sentiment = sum(sentiment_scores) / len(sentiment_scores)
        average_aspect_sentiments[aspect] = average_sentiment

for aspect, sentiment_score in average_aspect_sentiments.items():
    print(f"Aspect: {aspect}, Average Sentiment Score: {sentiment_score:.2f}")
```
