```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

# Sample reviews
reviews = [
    "good in this price",
    "Nice phone",
    "Thanks for flipkart but realme 5 phone in not value for money.",
    "Best proccesor",
    "Volume during Whatsapp calls are very low. No person can understand what they want to say",
    "Totally waste of money. It has hardware issues",
    "Absolutely good mobile in this price.Today 14th January 2022 still it's live with  best performance .",
    "Really awesome battery is good",
    "So good phone, camera is super, battery is super, value for money",
    "Nice product",
    "Happy with flipkart.. It's being a long time relationship with flipkart.. From 2019 to 2021..we are expecting more love and care of our products which we are ordering..",
    "The camera is much better but performance is no good if you want to purchase this phone then go with higher ram and storage",
    "very good productI have been using for 3 years nowRecommendedNo repairs till now Perfectly working Amazing cam",
    "Nice Product",
    "Very nice good looking mind-blowing phone",
    "i am writing this review after 2 years of usage .this phone has very less brightness.even in full brightness it seems like dull dull .performance of 3,32GB variant is very laggy .phone hangs so much .camera is average ,neither too good not too bad .battery is good .Looks of phone are good.",
    "4 years good work",
    "Happy",
    "Best",
    "Lags too much , camera not good.",
    "Best budget phone at this price range",
    "Feature loaded phone...",
    "Nice product and worth for money must buy it",
    "Hang out too many times" ,
    "Good phone for reasonable price. Don't expect too much." ,
    "Its a good device for little time users" ,
    "No comparison" ,
    "Nice quality" ,
    "Excellent phone" ,
    "Not good for heavy use and very long time" ,
    "Awesome product " ,
    "Not bad , value for money. Gud performance overall gud" ,
    "good" ,
    "Best phonThanku flipkart" ,
    "Battery backup is good & display is good and camera is good it's range for good quality" ,
    "I am writing this after 2.6 yrs. The battery performance is reduced now. Taking too much time to charge and battery is dropping faster than earlier. I am facing some storage issues and slowness.. other than this. Everything is looks so fine. Better front and back camera. Look is nice." ,
    "Love it. U can buy it best in year 2018 mid budget segment." ,
    "It's so wonderful product" ,
    "Most powerful divce under this prise rate.... " ,
    "I had purchased it in 2019 but I am using this still nowsuperb Mobile battery and camera is too awesome " ,
    "Very bad experience nd very bad for Flipkart emi options Im not Suggest from everyone... this soo poor experience." ,
    "good quality" ,
    "Lovely phone" ,
    "I have good experience with this phone." ,
    "Average performance .......Lethargic after 6 months" ,
    "Good mobile" ,
    "Garbage product, Cheap quality sensors." ,
    "Very good product and very good service and very good delivery time and very good language" ,
    "Very very nice" ,
    "good product" ,
    "Best phone under 10k" ,
    "Very nice" ,
    "Awesome" ,
    "Very nice phone" ,
    "Very good product" ,
    "Good product." ,
    "Very nice product" ,
    "Average quality" ,
    "Thanks" ,
    "Good product" ,
    "Worst product! Kindly do not purchase" ,
    "I suggest every one please do not consider realme budget phone. Very bed experience" ,
    "Very nice and beautiful mobile phoneVery best budge phone" ,
    "It's a little heavy so didn't liked it that much." ,
    "My phone is 3years old" ,
    "Satisfied" ,
    "Nice products" ,
    "Very good mobile" ,
    "Very gud phone in this price segment" ,
    "It's a very good phone for anyone in the mid section income buyer recommend buying it" ,
    "Very good" ,
    "Best phone" ,
    "Good phone" ,
    "Ok" ,
    "Overall great phone according to prize. Working well. Good features" ,
    "Khub valo phone ata camera quality & colour too good ......thanku flipcart",
    "I don't believe it wow it's very  good" ,
    "All good" ,
    "worth" ,
    "Mobile speaker not working, it's just 1year, I did not expect this bad performance, behaviour for real me brand, it would be appriciable if I get good service or response from RealMe" ,
    "Nice product " ,
    "Awesome performance" ,
    "After 1year of usage, it's having lag & lag too much" ,
    "Awesome phone at this price range.... wonderful battery life, very good display, super speaker output.Phone is sturdy also, still now dropped from shirt pocket many times, no physical damage (I have additionally add tempered screen guard (those selling at footpath wala screen guard)" ,
    "Motherboard  very poor" ,
    "Best performance " ,
    "What a phone man " ,
    "Jj" ,
    "Nice product" ,
    " nice" ,
    "Always Good Very good camera according this price" ,
    "Very good product battery backup is must" ,
    "Best product I ever ordered best service best value for money" ,
    "this phone is very good i love it " ,
    "Best products" ,
    "Nice" ,
    "It's Good, I'm Happy to purchase..." ,
    "Best camera" ,
    "Nice and good product" ,
    "awesome",
    "Nice",
    "Nice mobile.Value for money.",
    "I like realme",
    "Top posts phone h",
    "Quality is not good",
    "good mobile in budget",
    "Not so great",
    "Good products",
    "Best mobile" ,
    "Good" ,
    "Superb" ,
    "Nice phone",
    "good mob."
]

# Initialize the SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()

# Aspects to analyze
aspects_to_analyze = ["phone","camera", "price", "battery", "performance","realme"]

# Initialize a dictionary to store sentiment scores for each aspect
aspect_sentiments = {aspect: [] for aspect in aspects_to_analyze}

# Perform aspect-based sentiment analysis
for review in reviews:
    for aspect in aspects_to_analyze:
        if aspect in review.lower():
            sentiment_score = sia.polarity_scores(review)
            aspect_sentiments[aspect].append(sentiment_score['compound'])

# Display the sentiment scores for each aspect
for aspect, scores in aspect_sentiments.items():
    print(f"Aspect: {aspect}")
    for i, score in enumerate(scores, 1):
        print(f"Review {i}: Sentiment Score = {score:.2f}")
    print()
```

```python
average_aspect_sentiments = {}
for aspect, sentiment_scores in aspect_sentiments.items():
    if sentiment_scores:
        average_sentiment = sum(sentiment_scores) / len(sentiment_scores)
        average_aspect_sentiments[aspect] = average_sentiment

for aspect, sentiment_score in average_aspect_sentiments.items():
    print(f"Aspect: {aspect}, Average Sentiment Score: {sentiment_score:.2f}")
```
